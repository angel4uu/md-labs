# Análisis de componentes principales 
sonar.csv


## Librerias

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

## Datos

```python
df = pd.read_csv("sonar.csv")
```

```python
df.head()
```

```python
df.describe()
```

## Escalado

```python
from sklearn.preprocessing import StandardScaler
```

Dividar datos de target

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

Escalado de datos


```python
scaler = StandardScaler()
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
X_scaled
```

## PCA

Hallar coponentes

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = np.cumsum(explained_variance)
```

```python
var_df = pd.DataFrame({
    'Componente': np.arange(1, len(explained_variance) + 1),
    'Varianza_Explicada': explained_variance,
    'Varianza_Acumulada': cumulative_variance
})

var_df
```

Hallar número mínimo de componentes para 93% de varianza explicada.

```python
threshold = 0.93
n_components = np.argmax(cumulative_variance >= threshold) + 1
print(f"\nNúmero mínimo de componentes que explican al menos el 93% de la varianza: {n_components}")
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o', linestyle='--')
plt.axhline(y=threshold, color='r', linestyle='--', label='93% Varianza')
plt.axvline(x=n_components, color='g', linestyle='--', label=f'{n_components} Componentes')
plt.title('Varianza Acumulada por Número de Componentes')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Acumulada')
plt.legend()
plt.grid(True)
plt.show()
```

PCA con componentes óptimos

```python
pca_final = PCA(n_components=n_components)
X_pca_final = pca_final.fit_transform(X_scaled)
```

Dataframe con componentes óptimos

```python
cols = [f"PC{i+1}" for i in range(n_components)]
X_pca_df = pd.DataFrame(X_pca_final, columns=cols)

if y is not None:
    X_pca_df['Etiqueta'] = y.values

X_pca_df
```

Generar nuevo csv


```python
X_pca_df.to_csv("sonar_pca.csv", index=False)
print("\nArchivo 'sonar_pca.csv' guardado correctamente.")
```
