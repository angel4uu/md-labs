```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt 

# Datos de ejemplo
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([1, 4, 9, 16, 25])

# Ajustar el modelo de regresión lineal simple
model_simple = LinearRegression()
model_simple.fit(X, y)

# Predicciones
y_pred_simple = model_simple.predict(X)

# Visualización
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred_simple, color='red', label='Lineal Simple')
plt.xlabel('Variable Independiente (X)')
plt.ylabel('Variable Dependiente (y)')
plt.title('Comparación de Regresión Lineal Simple y Polinómica')
plt.legend()
plt.show()
```

```python
# Crear características polinómicas de grado 2
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Ajustar el modelo de regresión polinómica
model_poly = LinearRegression()
model_poly.fit(X_poly, y)

# Predicciones
y_pred_poly = model_poly.predict(X_poly)

# Visualización
plt.scatter(X, y, color='blue')
plt.plot(X, y_pred_simple, color='red', label='Lineal Simple')
plt.plot(X, y_pred_poly, color='green', linestyle='--', label='Polinómica Grado 2')
plt.xlabel('Variable Independiente (X)')
plt.ylabel('Variable Dependiente (y)')
plt.title('Comparación de Regresión Lineal Simple y Polinómica')
plt.legend()
plt.show()
```
